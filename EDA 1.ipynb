```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df = pd.merge(df,df_country,on = "Country Code",how = 'left')
```

```python
Cname = final_df.Country.value_counts().index
```

```python
Counts = final_df.Country.value_counts().values
```

```python
plt.pie(Counts[:3], labels = Cname[:3],autopct = '%1.2f%%')
```

Observation: Maximum outlets in India folllowed by US and UK.

```python
final_df.columns
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observations:
    
1. Ratings 4.5 - 4.9 indicates Excellent
2. Ratings 4.0 - 4.4 indicates Very Good
and so on from above table.

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12,8)
sns.barplot('Aggregate rating','Rating Count',hue = 'Rating color',data = ratings,palette = ['blue','red','orange','yellow','green','green'])
```

Observations:
    
1. Not rated restaurants counts are high.
2. Max rated restaurants ratings are between 4.5 - 4.9.

```python
sns.countplot(x = 'Rating color', data = ratings,palette = ['blue','red','orange','yellow','green','green'] )
```
